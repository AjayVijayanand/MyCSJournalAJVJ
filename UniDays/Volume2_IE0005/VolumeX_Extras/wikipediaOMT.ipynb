## 2016 Summer Olympics medal table

Link: https://en.wikipedia.org/wiki/2016_Summer_Olympics_medal_table

```python
import numpy as np
import pandas as pd
```

Import the Wikipedia page in Jupyter Notebook (check M1 DataAcquisition.ipynb for hints about this).

```python
OlympicsWP = pd.read_html("https://en.wikipedia.org/wiki/2016_Summer_Olympics_medal_table")
```

How many tables are in this Wikipedia page? Check the “len” of the imported page to find this out.

```python
print("Data type : ", type(OlympicsWP))
print("HTML tables : ", len(OlympicsWP))
```

Which one is the actual “2016 Summer Olympics medal table”? Explore all tables in the data to know.

```python
OlympicsMT = pd.read_html("https://en.wikipedia.org/wiki/2016_Summer_Olympics_medal_table", match='2016 Summer Olympics medal table')
OlympicsMT[0].head()
```

Store the main table, that is, “2016 Summer Olympics medal table”, as a new Pandas DataFrame.

```python
OlympicDict = pd.DataFrame(OlympicsMT[0])
OlympicDict
```

Extract the TOP 20 countries from the medal table, and store these rows as a new DataFrame.

#### Based on No. of Gold + Total Medals

```python
OlympicDict2 = OlympicDict.iloc[:20]
OlympicDict2
```

#### Based on Total ONLY

```python
OlympicDict3 = OlympicDict.sort_values(by='Total', ascending=False)
OlympicDict3
```

```python
OlympicDict4 = OlympicDict3.iloc[1:21]
OlympicDict4.reset_index()
```

Extra Things which aren't required but I wanna see it
