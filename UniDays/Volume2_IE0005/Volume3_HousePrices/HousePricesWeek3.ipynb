```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
sb.set()
```

Import the “train.csv” data from the downloaded data folder (has four files) in Jupyter Notebook.

```python
houseprice = pd.read_csv('train.csv')
```

What are the data types (“dtypes”) – Numeric/Categorical – of the variables (columns) in the dataset?

```python
print(houseprice.dtypes)
```

Extract only the variables (columns) of type Integer (int64), and store as a new Pandas DataFrame.

```python
int_houseprice = pd.DataFrame(houseprice.select_dtypes('int64'))
```

Drop non-Numeric variables from the DataFrame to have a clean DataFrame with Numeric variables

```python
clean_houseprice = int_houseprice.drop(columns=['Id', 'MSSubClass', 'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'LowQualFinSF', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath', 'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces', 'GarageCars', 'MoSold', 'YrSold'], axis='1')
```

Find the Summary Statistics (Mean, Median, Quartiles etc) of SalePrice from the Numeric DataFrame.

```python
SalesPrices = pd.DataFrame(clean_houseprice['SalePrice'])
SalesPrices.describe()
```

```python
g, axes = plt.subplots(3, 1, figsize=(20, 10))
sb.boxplot(data = SalesPrices, orient = "h", ax = axes[0])
sb.histplot(data = SalesPrices, ax = axes[1])
sb.kdeplot(data = SalesPrices, ax = axes[2])
```

Find the Summary Statistics (Mean, Median, Quartiles etc) of LotArea from the Numeric DataFrame.

```python
LotAreas = pd.DataFrame(clean_houseprice['LotArea'])
LotAreas.describe()
```

```python
f, axes = plt.subplots(3, 1, figsize=(20, 10))
sb.boxplot(data = LotAreas, orient = "h", ax = axes[0])
sb.histplot(data = LotAreas, ax = axes[1])
sb.kdeplot(data = LotAreas, ax = axes[2])
```

Plot SalePrice (y-axis) vs LotArea (x-axis) using jointplot.

```python
PriceVArea = pd.concat([SalesPrices, LotAreas], axis = 1)
sb.jointplot(data = PriceVArea, x = "LotArea", y = "SalePrice", height = 12)
```
